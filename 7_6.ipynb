병합

```python
import pandas as pd
import numpy as np

d1 = {
    '기준' : [0, 1, 2, 3, 4, 5],
    "data1" : [4, 5, 2, 3, 4, 5],
    'data2' : ['A', 'B', 'D', 'E', 'F', 'Z']
}
df1 = pd.DataFrame(d1,columns = ['기준','data1','data2'])
df1
```

```python
d2 = {
    '기준' : [2, 4, 9, 3],
    "data1" : [4, 5, 2, 3],
    'data3' : ['data1', 'data2', 'data3', 'data4']
}
df2 = pd.DataFrame(d2,columns = ['기준','data1','data3'])
df2
```

```python
df1, df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on="기준")
```

```python
pd.merge(df1, df2, how = "outer", on = "기준")
```

```python
pd.merge(df1, df2, how = "left", on = "기준")
```

```python
pd.merge(df1, df2, how = "right", on = "기준")
```

```python
from numpy.linalg import inv

data1 = [[2, 3], [1, -2]]
data2 = [[1], [4]]

arr1 = np.array(data1)
arr2 = np.array(data2)
result = inv(arr1)@arr2
x = result[0, 0]
y = result[1, 0]
x, y
```

```python
2x + 3y = 1
x - 2y = 4
```
