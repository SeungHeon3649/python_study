```python
# 필수 라이브러리
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
# 전체 데이터셋 읽어들이기
raw_df = pd.read_csv('./data/continuous_factory_process.csv', index_col="time_stamp")
submission = pd.read_csv('data/submission_data.csv')
```

```python
# setpoint 열 지우기
setpoint_col = [col for col in raw_df.columns if 'Setpoint' in col]
raw_df.drop(setpoint_col, axis=1, inplace=True)

# stage1 output
stage1_output_col = [col for col in raw_df.columns if 'Stage1.Output' in col]
stage1_answer = raw_df.loc[:, stage1_output_col]

# stage2 output
stage2_output_col = [col for col in raw_df.columns if 'Stage2.Output' in col]
stage2_answer = raw_df.loc[:, stage2_output_col]

# feature data
features1 = raw_df.columns[:41]
features2 = raw_df.columns[56:70]
```

```python
# model
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model2 = LinearRegression()
```

```python
# stage1
model1.fit(raw_df[features1], raw_df[stage1_output_col])
submission1 = model1.predict(submission[features1])
np.save('submission1.npy', submission1)
```

```python
# stage2
model2.fit(raw_df[features2], raw_df[stage2_output_col])
submission2 = model2.predict(submission[features2])
np.save('submission2.npy', submission2)
```
