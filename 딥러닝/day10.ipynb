```python
from keras.datasets import mnist
import numpy as np
(x_data,y_data),(tt_x,tt_y) = mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split
tr_x,val_x,tr_y,val_y = train_test_split(x_data,y_data,test_size=0.2,stratify = y_data)
```

```python
from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout, BatchNormalization
```

```python
from keras.optimizers import SGD
from keras.losses import sparse_categorical_crossentropy,categorical_crossentropy
```

```python
sgd = SGD(0.001)
```

```python
m = Sequential()
m.add(Flatten(input_shape=tr_x.shape[1:]))
m.add(Dense(50,activation='sigmoid')) # 50이라는 뉴런수만 주면 됨
m.add(Dense(50,activation='sigmoid'))
m.add(Dense(50,activation='sigmoid'))
m.add(Dense(50,activation='sigmoid'))
m.add(Dense(50,activation='sigmoid'))
m.add(Dense(50,activation='sigmoid'))
m.add(Dense(10,activation='softmax'))
m.compile(optimizer=sgd, loss='sparse_categorical_crossentropy', metrics = 'acc')
m.summary()
```

```python
hy = m.fit(tr_x,tr_y,validation_data=(val_x,val_y),epochs=100,verbose=0)
```

```python
import matplotlib.pyplot as plt
plt.plot(hy.history['loss'])
plt.plot(hy.history['val_loss'])
plt.show()
plt.plot(hy.history['acc'])
plt.plot(hy.history['val_acc'])
plt.show()
```

```python
import pandas as pd
df = pd.read_table('ratings_test.txt', encoding = 'utf-8')[['document', 'label']]
```

```python
from konlpy.tag import Okt
okt = Okt() # 형태소 분석기
```

```python
df.info()
```

```python
# 1. 결측치 제거
df = df.dropna()
```

```python
# 2. w중복 제거
df = df.drop_duplicates(subset = ['document'])
```

```python
import numpy as np
df.document = df.document.str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣]', '') # 한글 제외 단어 삭제
df.document = df.document.str.replace('^ +', '')
df.document = df.document.replace('', np.nan)
df = df.dropna()
```

```python
df.info()
```

```python
s_w = set(['은', '는', '이', '가', '를', '들', '에게', '의', '도', '을', '으로', '만','라서', '하다'])
```

```python
from tqdm import tqdm
X_data = []
for i in tqdm(df['document']):
    tk_d = okt.morphs(i)
    end_d = [w for w in tk_d if w not in s_w]
    X_data.append(' '.join(end_d))
```

```python
y_data = df.label.values
```

```python
len(X_data), len(y_data)
```

```python
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y =  train_test_split(X_data, y_data, test_size = 0.3, stratify = y_data)
```

```python
tk = Tokenizer()
tk.fit_on_texts(train_x)
```

```python
n =len([d for d in sorted(list(tk.word_counts.items()), key = lambda x:x[1]) if d[1] >4])
```

```python
token = Tokenizer(n)
token.fit_on_texts(train_x)
```

```python
token_train_x = token.texts_to_sequences(train_x)
token_test_x = token.texts_to_sequences(test_x)
```

```python
train_inputs = pad_sequences(token_train_x)
test_inputs = pad_sequences(token_test_x)
```
