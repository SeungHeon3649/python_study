```python
class 계좌관리:
  총계좌정보={}
  def __init__(self,고객명,계좌번호,금액):
    self.고객명=고객명
    self.계좌번호=계좌번호
    self.금액=금액
    self.통장내역=f'{"초기금액:"+str(금액):10s}|{"합계금액:"+str(금액):10s}\n'

  def __str__(self):
    return f"|{self.계좌번호:^11}|{self.고객명:^5}|{self.금액:>10}|"
  def 내역기록(self,금액):
    if 금액>=0:
      self.통장내역 += f'{"입금  액:"+str(금액):10s}|{"합계금액:"+str(self.금액):10s}\n'
    else:
      self.통장내역 += f'{"출금  액:"+str(금액* -1):10s}|{"합계금액:"+str(self.금액):10s}\n'

  @classmethod
  def 생성(cls):
    print(f"""--------------
계좌생성
--------------""")
    계좌번호=input("계좌번호: ")
    고객명=input("계좌주: ")
    금액=int(input("초기입금액: "))
    n_통장=cls(고객명,계좌번호,금액)
    if 계좌번호 in cls.총계좌정보:
      print("중복된 계좌 번호 입니다.")
      return
    cls.총계좌정보[계좌번호]=n_통장
    print("결과: 계좌가 생성되었습니다.")
  @classmethod
  def 출력(cls):
    print(f"""--------------
계좌목록
--------------""")
    for 통장 in cls.총계좌정보.values():
      print(통장)
  @classmethod
  def 입금(cls):
    print(f"""--------------
예금
--------------""")
    입금계좌=input("계좌번호: ")
    입금액=int(input("예금액: "))
    if 입금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    호출통장=cls.총계좌정보[입금계좌]
    호출통장.금액+=입금액
    호출통장.내역기록(입금액)
    print('결과: 예금이 성공되었습니다.')
  @classmethod
  def 출금(cls):
    print(f"""--------------
출금
--------------""")
    출금계좌=input("계좌번호: ")
    출금액=int(input("금액: "))
    if 출금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    호출통장=cls.총계좌정보[출금계좌]
    if 호출통장.금액<출금액:
      print("잔액 부족")
      return
    호출통장.금액-=출금액
    출금액*=-1
    호출통장.내역기록(출금액)
    print('결과: 출금이 성공되었습니다.')
  @classmethod
  def 종료(cls):
     print("프로그램 종료")
     return True
  @classmethod
  def 내역_조회(cls):
    print(f"""--------------
내역조회
--------------""")
    확인계좌=input("계좌번호: ")
    if 확인계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    print(f"{확인계좌}의 내역")
    print(cls.총계좌정보[확인계좌].통장내역)
    print("내역 출력이 완료되었습니다.")
  @classmethod
  def 송금(cls):
    print(f"""--------------
송금
--------------""")
    출금계좌=input("출금 계좌번호: ")
    if 출금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    입금계좌=input("입금 계좌번호: ")
    if 입금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    출금액=int(input("송금액: "))
    출금통장=cls.총계좌정보[출금계좌]
    입금통장=cls.총계좌정보[입금계좌]
    if 출금통장.금액<출금액:
      print("잔액 부족")
      return
    출금통장.금액-=출금액
    입금액=출금액
    출금액*=-1
    출금통장.내역기록(출금액)
    입금통장.금액+=입금액
    입금통장.내역기록(입금액)
    print("송금 완료")
```

```python
import pickle
ui=f"""------------------------------------------------------------------
1.계좌생성 | 2.계좌목록 | 3.예금 | 4.출금 | 5.종료 | 6.내역조회 | 7.송금
------------------------------------------------------------------
선택>"""
계좌프로그램=[계좌관리.생성,계좌관리.출력,계좌관리.입금,
        계좌관리.출금,계좌관리.종료,계좌관리.내역_조회,계좌관리.송금]
try:
  계좌관리.총계좌정보=pickle.load(open("save_data.p",'rb'))
except:
  계좌관리.총계좌정보={}

def main():
  try:
    m=int(input(ui)[0])-1
  except:
    print(f"잘못 입력되었습니다.\n입력은 1~{len(계좌프로그램)}까지만 가능합니다.")
    return
  else:
    if not (0<=m<=len(계좌프로그램)-1):
      print(f"잘못 입력되었습니다.\n입력은 1~{len(계좌프로그램)}까지만 가능합니다.")
      return
    return 계좌프로그램[m]()

if __name__=="__main__":
  run=False
  while not run:
    run=main()
  else:
    pickle.dump(계좌관리.총계좌정보,open("save_data.p",'wb'))
```
