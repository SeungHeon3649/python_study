```python
def 삽입정렬(입력데이터, 모드 = 0):
    정렬결과=[]
    #1. 원소추출
    while 입력데이터:
        추출값 = 입력데이터.pop(0)
        n=len(정렬결과)-1
    #2. 비교
        while True:
            if len(정렬결과)==0 or n==-1:
                비교자리=0
                break
            elif 모드==0 and 정렬결과[n]<추출값:
                비교자리=n+1
                break
            elif 모드==1 and 정렬결과[n]>추출값:
                비교자리=n+1
                break
            else:
                n-=1
    #3. 원소 삽입
        정렬결과.insert(비교자리,추출값)
    return 정렬결과

l = [2, 4, 1, 6, 3, 9]
삽입정렬(l)
```

```python
def 쉘(l, h):
    ck_l = [[] for _ in range(h)]
    for i in range(len(ck_l)):
        st = i
        add_idx = st
        while add_idx < len(l):
            ck_l[i].append(l[add_idx])
            add_idx += h
        ck_l[i] = 삽입정렬(ck_l[i])

    # 삽입
    end_a=[None for _ in range(len(l))]
    for i in range(len(ck_l)):
        for j in range(len(ck_l[i])):
            idx_i= i + (j * h)
            end_a[idx_i] = ck_l[i][j]
    h //= 2
    return end_a, h

def 쉘_정렬(l, h):
    l, h = 쉘(l, h)
    while True:
        if h != 0:
            l, h = 쉘(l, h)
        else:
            return l

l = [6, 25, 13, 56, 2, 3, 71, 93, 1, 7, 5, 8]
h = 4
end_l = 쉘_정렬(l, h)
print(end_l)
```

```python
from random import randint, sample
import time
import matplotlib.pyplot as plt

l = sample(range(1,101), 100)
```

```python
def 쉘_정렬(in_l):
    l = in_l[:]
    h = 1
    while h < len(l):
        h = (3 * h) + 1
    h //= 3
    while h >= 1:
        for i in range(h, len(l)):
            j = i
            while j >= h and l[j] < l[j - h]:
                l[j], l[j - h] = l[j - h], l[j]
                j -= h
        h //= 3
        plt.bar(range(len(l)), l, width = 1)
        plt.show()
    return l
end_l = 쉘_정렬(l)
print(end_l)
```
