```python
class 자동차:
  UI=f"""### 자동차 정보 프로그램 ####
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 종료
{'#'*30}
입력:"""
  자동차_리스트=[]
  최고속도_v=0
  최저속도_v=1000
  def __init__(self,이름,엔진가격,타이어가격,옵션유무,옵션이름,최고속도,구매가격):
    self.이름=이름
    self.엔진가격=엔진가격
    self.타이어가격=타이어가격
    self.옵션유무=옵션유무
    self.옵션이름=옵션이름
    self.최고속도=최고속도
    self.구매가격=구매가격
  def __str__(self):
    return f"""{'-'*30}
이름:{self.이름}
엔진가격:{self.엔진가격}
타이어가격:{self.타이어가격}
옵션유무:{self.옵션유무}
옵션이름:{self.옵션이름}
최고속도:{self.최고속도}
구매가격:{self.구매가격}
{'-'*30}
"""
  def 구매가격출력(self):
    return f"""{'-'*30}
이름:{self.이름}
구매가격:{self.구매가격}
{'-'*30}
"""
  @classmethod
  def 정보입력(cls):
    이름=input("이름:")
    엔진가격=int(input("엔진가격:"))
    타이어가격=int(input("타이어가격:"))
    옵션유무=input("옵션유무(유/무):")
    if 옵션유무=='유':
      옵션이름=input("옵션이름:")
    else:
      옵션이름="없음"
    최고속도=int(input("최고속도:"))
    cls.최고속도_v=cls.최고속도_v if cls.최고속도_v>최고속도 else 최고속도
    cls.최저속도_v=cls.최저속도_v if cls.최저속도_v<최고속도 else 최고속도
    구매가격=엔진가격+타이어가격
    cls.자동차_리스트.append(cls(이름, 엔진가격, 타이어가격, 옵션유무, 옵션이름, 최고속도, 구매가격))
  @classmethod
  def 정보_총출력(cls):
    print("자동차 정보 출력")
    print("-"*20)
    for 자동차 in cls.자동차_리스트:
      print(자동차)
    else:
      print("정보 조회가 완료 되었습니다.")
      input()
  @classmethod
  def 구매가격_출력(cls):
    print("구매가격 정보 출력")
    print("-"*20)
    for 자동차 in cls.자동차_리스트:
      print(자동차.구매가격출력())
    else:
      print("정보 조회가 완료 되었습니다.")
      input()
  @classmethod
  def 옵션_정보출력(cls):
    print("자동차 정보 출력")
    print("-"*20)
    for 자동차 in cls.자동차_리스트:
      if 자동차.옵션유무=='유':
        print(자동차)
    else:
      print("정보 조회가 완료 되었습니다.")
      input()
  @classmethod
  def 속도차_출력(cls):
    print(f"기록된 차량의 속도차는{cls.최고속도_v-cls.최저속도_v}입니다.")
```

```python
def main():
  매뉴=int(input(자동차.UI))
  if 매뉴==1:
    자동차.정보입력()
  elif 매뉴==2:
    자동차.정보_총출력()
  elif 매뉴==3:
    자동차.구매가격_출력()
  elif 매뉴==4:
    자동차.옵션_정보출력()
  elif 매뉴==5:
    자동차.속도차_출력()
  elif 매뉴==6:
    return True
if __name__=="__main__":
  run=False
  while not run:
    run=main()
  else:
    print("프로그램이 종료됩니다.")
```

```python
class 자동차:
  UI=f"""### 자동차 정보 프로그램 ####
1. 자동차 정보 입력
2. 저장된 목록 보기
3. 각 자동차의 구매 가격 조회
4. 옵션이 있는 자동차 정보 조회
5. 빠른 자동차와 느린 자동차의 속도차 조회
6. 종료
{'#'*30}
입력:"""
  자동차_리스트=[]
  최고속도_v=0
  최저속도_v=1000
  def __init__(self,이름,엔진가격,타이어가격,옵션유무,옵션이름,최고속도,구매가격):
    self.이름=이름
    self.엔진가격=엔진가격
    self.타이어가격=타이어가격
    self.옵션유무=옵션유무
    self.옵션이름=옵션이름
    self.최고속도=최고속도
    self.구매가격=구매가격
  def __repr__(self):
    return f"""{'-'*30}
이름:{self.이름}
엔진가격:{self.엔진가격}
타이어가격:{self.타이어가격}
옵션유무:{self.옵션유무}
옵션이름:{self.옵션이름}
최고속도:{self.최고속도}
구매가격:{self.구매가격}
{'-'*30}
"""
  def 구매가격출력(self):
    return f"""{'-'*30}
이름:{self.이름}
구매가격:{self.구매가격}
{'-'*30}
"""
  @classmethod
  def 정보입력(cls):
    이름=input("이름:")
    엔진가격=int(input("엔진가격:"))
    타이어가격=int(input("타이어가격:"))
    옵션유무=input("옵션유무(유/무):")
    if 옵션유무=='유':
      옵션이름=input("옵션이름:")
    else:
      옵션이름="없음"
    최고속도=int(input("최고속도:"))
    cls.최고속도_v=cls.최고속도_v if cls.최고속도_v>최고속도 else 최고속도
    cls.최저속도_v=cls.최저속도_v if cls.최저속도_v<최고속도 else 최고속도
    구매가격=엔진가격+타이어가격
    cls.자동차_리스트.append(cls(이름, 엔진가격, 타이어가격, 옵션유무, 옵션이름, 최고속도, 구매가격))
  @classmethod
  def 정보_총출력(cls):
    print("자동차 정보 출력")
    print("-"*20)
    for 자동차 in cls.자동차_리스트:
      print(자동차)
    else:
      print("정보 조회가 완료 되었습니다.")
      input()
  @classmethod
  def 구매가격_출력(cls):
    print("구매가격 정보 출력")
    print("-"*20)
    for 자동차 in cls.자동차_리스트:
      print(자동차.구매가격출력())
    else:
      print("정보 조회가 완료 되었습니다.")
      input()
  @classmethod
  def 옵션_정보출력(cls):
    print("자동차 정보 출력")
    print("-"*20)
    for 자동차 in cls.자동차_리스트:
      if 자동차.옵션유무=='유':
        print(자동차)
    else:
      print("정보 조회가 완료 되었습니다.")
      input()
  @classmethod
  def 속도차_출력(cls):
    print(f"기록된 차량의 속도차는{cls.최고속도_v-cls.최저속도_v}입니다.")
```

```python
import pickle
def main():
  매뉴=int(input(자동차.UI))
  if 매뉴==1:
    자동차.정보입력()
  elif 매뉴==2:
    자동차.정보_총출력()
  elif 매뉴==3:
    자동차.구매가격_출력()
  elif 매뉴==4:
    자동차.옵션_정보출력()
  elif 매뉴==5:
    자동차.속도차_출력()
  elif 매뉴==6:
    return True
if __name__=="__main__":
  run=False
  try:
    자동차.자동차_리스트=pickle.load(open("자동차_data.p",'rb'))
  except:
    print("파일이 없습니다.")
  else:
    print("파일을 로드 했습니다.")
  while not run:
    run=main()
  else:
    with open("자동차_data.p",'wb') as f:
      pickle.dump(자동차.자동차_리스트,f)
    print("프로그램이 종료됩니다.")
```
