#객체지향


```python
# class 클래스명
class A:
    이름 = ""
    번호 = 0

a1 = A()
a2 = A()
a1.이름 = "소방차"
a1.번호 = 1234
a2.이름 = "구급차"
a2.번호 = 1004
a1.이름, a2.이름, a1.번호, a2.번호
```

```python
class A:
    def __init__(self, 이름, 번호):
        self.이름 = 이름
        self.번호 = 번호

a1 = A("소방차", 1234)
a1.이름, a1.번호
```

```python
class A:
    클래스필드 = "클래스필드 내용"
```

```python
A.클래스필드
```

```python
a1 = A()
a2 = A()
print(a1.클래스필드)
a1.클래스필드 = 10
print(a1.클래스필드)
```

```python
class 메서드:
    def m1(self):
        print("m1호출")
    def m2(self, 변수):
        print(변수, "m2호출")
    def m3(self):
        print("m3호출")
    def m4(self, 변수1, 변수2):
        print(변수1, 변수2, "m4호출")
    def m5(self, 변수1):
        return f"{변수1} m5호출"
```

```python
m_all = 메서드()
m_all.m1()
m_all.m2(1)
m_all.m3()
m_all.m4(1, 2)
m_all.m5(1)
```
