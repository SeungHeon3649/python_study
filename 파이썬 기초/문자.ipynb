Q3 지정된 편입 종목 상위 데이터를 모두 정리하여 csv파일로 완성하시오

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
class 테이블_스크래핑:
  def __init__(self,기준주소):
    self.기준주소=기준주소
    self.end_data=[]
    self.제목=None
  def 변환함수(self,data):
    return data.replace('\n','').replace('\t','').replace(',','')
  def 제목추출(self):
    최종정보=self.기준주소
    html=requests.get(최종정보).text
    bs=BeautifulSoup(html,'lxml')
    제목=[i.get_text() for i in bs.select('table.type_1 tr')[0].select('th')]
    return 제목
  def 수집(self,숫자):
    최종정보=self.기준주소+str(숫자)
    html=requests.get(최종정보).text
    bs=BeautifulSoup(html,'lxml')
    for d in bs.select('table.type_1 tr')[1:]:
      if not d.get_text():
        continue
      data=[]
      for i in d.select('td'):
        data.append(self.변환함수(i.get_text()))
      self.end_data.append(data)
    time.sleep(1)
    print(숫자,"회 수집",sep="")
  def 동작(self,횟수):
    self.제목=self.제목추출()
    for i in range(1,횟수+1):
      self.수집(i)
    self.저장()
  def 저장(self):
    df=pd.DataFrame(self.end_data,columns=self.제목)
    df.to_csv("data1.csv",index=False, encoding='utf-8')
    df.to_csv("data2.csv",index=False, encoding='cp949')
if __name__=='__main__':
  스크래핑=테이블_스크래핑('https://finance.naver.com/sise/entryJongmok.naver?&page=')
  스크래핑.동작(20)
```
