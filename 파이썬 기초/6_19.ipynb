```python
def 리스트_출력(st):
  print('[',end=" ")
  while st:
    if not st.next:
      print(st.data,end=" ")
    else:
      print(st.data,end=" -> ")
    st=st.next
  else:
    print("]")
    print("리스트의 출력이 종료 되었습니다.")
def 끝에서_삽입(st,x):
  #빈 리스트 확인
  if not st:
    st=Node(x,None)
  else:
    nn=st
    while nn.next:
      nn=nn.next
    else:
      nn.next=Node(x,None)
  return st
def 중간_삽입(st,n,x):
  #비어있는 삽입
  if not st:
    st=Node(x,None)
    return st
  #머리에서 삽입
  if n==1:
    st=Node(x,st)
    return st
  #n번째 노드 위치에서 삽입(n-1의 위치 노드)
  p=st#1번째 노드
  이동수=0
  #삽입전 노드위치 찾기
  while True:
    #이동 종료
    if 이동수==n-2:
      break
    p=p.next
    이동수+=1
  #1.데이터를 가지고있는 노드 생성
  #next_n=Node(x)
  #2.전 노드의 link연결
  #next_n.next=p.next
  #3.생성노드와 전노드 연결
  #p.next=next_n
  p.next=Node(x,p.next)
  return st
```

삭제
- 빈곳 - 절대 불가
- 끝 - 끝노드 제거
- 머리(중간) - 삭제 노드 뒷노드 앞단 연결

```python
def 끝_삭제(st):
    if st == None:
        print("빈 리스트는 삭제 불가")
        return
    t_p = st
    while t_p.next:
        t_p = t_p.next
    t_p.next = None

def 중간_삭제(st, n):
    if st == None:
        print("빈 리스트는 삭제 불가")
        return
    if n == 0:
        st = st.next
        return st
    else:
        if n == 1:
            st.next = st.next.next
        t_p = st # 1번째 노드
        for i in range(n - 1):
            t_p = t_p.next
        t_p.next = t_p.next.next
```
