```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("data.csv")
X=df[['D_length',	'D_weight']]
y=df['y']
X.shape,y.shape
```

```python
s=StandardScaler().fit(X)
s_X=s.transform(X)
```

```python
tr_x,tt_x,tr_y,tt_y=train_test_split(s_X,y,random_state=1,test_size=0.3)
```

```python
tr_x.shape,tt_x.shape,tr_y.shape,tt_y.shape
```

```python
m=LogisticRegression()
m.fit(tr_x,tr_y)
```

```python
m.score(tr_x,tr_y),m.score(tt_x,tt_y)
```

```python
m.coef_,m.intercept_
```

```python
from sklearn.metrics import f1_score,accuracy_score,precision_score,recall_score
```

```python
py=m.predict(tt_x)
```

```python
f1_score(tt_y,py)
```

```python
print(f"""정확도:{accuracy_score(tt_y,py)}
정밀도:{precision_score(tt_y,py)}
민감도:{recall_score(tt_y,py)}
f1점수:{f1_score(tt_y,py)}""")
```

```python
from sklearn.metrics import classification_report
print(classification_report(tt_y,py))
```
