```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("data.csv")
X = df[['D_length', "D_weight"]]
y = df['y']
X.shape, y.shape
```

```python
s = StandardScaler().fit(X)
s_X = s.transform(X)
```

```python
train_x, test_x, train_y, test_y = train_test_split(s_X, y, random_state = 1, test_size = 0.3)
```

```python
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
m = LogisticRegression()
m.fit(train_x, train_y)
```

```python
m.score(train_x, train_y), m.score(test_x, test_y)
```

```python
m.coef_, m.intercept_
```

```python
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
```

```python
py = m.predict(test_x)
```

```python
f1_score(test_y, py)
```

```python
print(f"""정확도 : {accuracy_score(test_y, py)}
정밀도 : {precision_score(test_y, py)}
민감도 : {recall_score(test_y, py)}
f1점수 : {f1_score(test_y, py)}""")
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y, py))
```

Q1 타이타닉 데이터를 이용하여 이진 분류기를 설계 하시오

1. 데이터 수집

```python
import seaborn as sns
df=sns.load_dataset('titanic')
df
```

2. 데이터 전처리
 1. 결측치 처리
 2. 이상치 처리

```python
df.info()
```

-int64형 칼럼
1. PassengerId: 탑승자 데이터 일련번호
2. Survived: 생존 여부 (0=사망, 1=생존)
3. Pclass: 티켓의 선실 등급 (1=일등석, 2=이등석, 3=삼등석)
4. SibSp: 같이 탑승한 형제자매 또는 배우자 인원수
5. Parch: 같이 탑승한 부모님 또는 어린이 인원수
-float64형 칼럼
1. Age: 탑승자 나이
2. Fare: 요금
-object(문자열)형 칼럼
1. Name: 탑승자 이름
2. Sex: 탑승자 성별
3. Ticket: 티켓 번호
4. Cabin: 선실 번호
5. Embarked: 중간 정착 항구 (C=Cherbourg, Q=Queenstown, S=Southampton)

```python
# 결측치 처리
df.drop(['deck'], axis = 1, inplace = True)
df.info()
```

```python
# age 결측치 평균치로 처리 & 결측치 버리기
df['age'] = df['age'].fillna(df.groupby('sex')['age'].transform('mean'))
df.dropna(subset = ['embark_town', 'embarked'], inplace = True)
df.isnull().sum()
```

```python
df.info()
```

```python
df
```

```python
# 중복느낌인 피처 제거
df = df.drop(['class', 'who', 'adult_male', 'embark_town', 'alive'], axis = 1)
df.info()
```

```python
df['alone']
```

```python
sc_df = df[['pclass', 'age', 'sibsp', 'parch', 'fare']]
ca_df = df[['sex', 'embarked', 'alone']]
y = df['survived']
```

3. 범주형처리

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe.fit(ca_df)
```

```python
ohe.feature_names_in_
```

```python
ohe.categories_
```

```python
ohe_ca_df = ohe.transform(ca_df)
ohe_ca_df.toarray().shape
```

```python
ohe_df = pd.DataFrame(ohe_ca_df.toarray(), columns = ohe.get_feature_names_out())
ohe_df
```

4. 범위처리

```python
from sklearn.preprocessing import MinMaxScaler
s = MinMaxScaler().fit(sc_df)
ss_df = s.transform(sc_df)
s_df = pd.DataFrame(ss_df, columns = sc_df.columns)
s_df
```

5. 전처리된 데이터 데이터 프레임 화

```python
end_df = pd.concat([ohe_df, s_df], axis = 1)
end_df
```

3.학습

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(end_df, y, random_state= 1, test_size = 0.3)
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
m = DecisionTreeClassifier()
m.fit(train_x, train_y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn_cm = KNeighborsClassifier().fit(train_x, train_y)
```

4.테스트 및 검증

```python
m.score(train_x, train_y), m.score(test_x, test_y)
```

```python
knn_cm.score(train_x, train_y), knn_cm.score(test_x, test_y)
```
