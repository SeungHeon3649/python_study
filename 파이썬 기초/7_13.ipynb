```python
def weather(area):
    url = 'https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q=' + area + '날씨'
    html = requests.get(url).text
    s_w = BeautifulSoup(html, 'lxml')
    l = s_w.select('dl.dl_weather')
    try:
        temp = s_w.select_one('strong.txt_temp').get_text()
        print(area, '온도 : ', temp, l[0].get_text(), l[1].get_text(), l[2].get_text())
    except:
        print("온도 가져오기 실패")

while True:
    area = input("지역을 입력하시오 : ")
    if area == '0':
        print("프로그램을 종료합니다.")
        break
    weather(area)
```

```python
# 안 알려줘도 되는 코드 (용주형네 크롤링)

from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
import pandas as pd
import time
import datetime


d = Chrome()
base_date = datetime.date(2015, 12, 31)
df_list = []

# 1000 x 1000크기로 창 크기 세팅
d.set_window_size(1000, 1000)

# 해당 페이지로 창을 연다
d.get('http://www.price.go.kr/tprice/portal/pricetrend/agriculturalprice/agriculturalPrice.do')
input_date = base_date.strftime('%Y-%m-%d')

# while input_date < str(datetime.datetime.now()):

base_date = base_date + datetime.timedelta(days=1)
input_date = base_date.strftime('%Y-%m-%d')


# 해당 xpath 정보를 input_t에 저장
input_t = d.find_element(By.XPATH, '//*[@id="regday"]')
# 입력된 값을 초기화
input_t.clear()
# 검색할 문자열 입력
input_t.send_keys(input_date)
time.sleep(1)

# 일자 선택창에서 닫기 버튼 누르기
b = d.find_element(By.XPATH, '//*[@id="ui-datepicker-div"]/div[2]/button[2]')
b.click()
time.sleep(1)

# 부류에서 채소류를 누르기
b = d.find_element(By.XPATH, '//*[@id="itemCategoryCode"]/option[3]')
b.click()
time.sleep(1)

# 품목에서 오이 누르기
b = d.find_element(By.XPATH, '//*[@id="itemCode"]/option[10]')
b.click()
time.sleep(1)

# 품종에서 가시 누르기
b = d.find_element(By.XPATH, '//*[@id="kindCode"]/option[3]')
b.click()
time.sleep(1)

# 조회 누르기
b = d.find_element(By.XPATH, '//*[@id="search_btn"]')
b.click()


# 현재 드라이버가 제어하고 있는 창에서 html을 가져온다
html = d.page_source

# html을 읽어와서 해당 페이지에 있는 table을 가져옴
df_l = pd.read_html(html)

df = df_l[1][4:]
df['날짜'] = input_date

# 데이터프레임 리스트에 추가
df_list.append(df)
```
