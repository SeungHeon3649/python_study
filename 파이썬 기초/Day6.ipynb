```python
#클래스 2개

class Account:

    account_l = []
    def __init__(self, in_account_num, in_account_name, in_money):
        self.account_num = in_account_num
        self.account_name = in_account_name
        self.money = in_money
        self.history_l = []

    def __repr__(self):
        return f"""{self.account_num}\t{self.account_name}\t{self.money}"""

    @classmethod
    def account_creation(cls):
        account_num = input("계좌번호: ")
        account_name = input("계좌주: ")
        money = int(input("초기입금액: "))
        #cls.account_l.append(cls(account_num, account_name, money))
        #print(account_list)
        try:
            cls.account_l.append(cls(account_num, account_name, money))
            print("결과: 계좌가 생성되었습니다.")
        except:
            print("계좌가 생성되지 않았습니다.")

    @classmethod
    def account_list(cls):  #계좌목록
        print("-" * 10)
        print("계좌목록")
        print("-" * 10)
        for account in cls.account_l:
            print(account)
        else:
            input()

    @classmethod    #예금
    def deposit(cls):
        print("-" * 10)
        print("예금")
        print("-" * 10)
        account_num = input("계좌번호: ")
        deposit_money = int(input("예금액: "))
        for account in cls.account_l:
            if account_num == account.account_num:
                print("결과: 예금이 성공되었습니다.")
                account.history_l.append(f"{deposit_money}원을 입금하였습니다.")
                account.money += deposit_money
                break

    @classmethod
    def whitdraw(cls):  #출금
        print("-" * 10)
        print("출금")
        print("-" * 10)
        account_num = input("계좌번호: ")
        whitdraw_money = int(input("출금액: "))
        for account in cls.account_l:
            if account_num == account.account_num:
                if account.money > whitdraw_money:
                    print("결과: 출금이 성공되었습니다.")
                    account.money -= whitdraw_money
                    account.history_l.append(f"{whitdraw_money}원을 출금하였습니다.")
                    break
                else:
                    print("결과: 출금 실패하였습니다.")

    @classmethod
    def remittance(cls):  #송금
        print("-" * 10)
        print("송금")
        print("-" * 10)
        account_num = input("계좌번호: ")
        for account in cls.account_l:
            if account_num == account.account_num:
                remittance_num = input("송금할 계좌번호: ")
                remittance_money = int(input("송금액: "))
                for account2 in cls.account_l:
                    if remittance_num == account2.account_num:
                        if account2.money > remittance_money:
                            print("결과 송금이 성공되었습니다.")
                            account.money -= remittance_money
                            account2.money += remittance_money
                            account.history_l.append(f"{remittance_num}에 {remittance_money}원을 송금하였습니다.")
                            account2.history_l.append(f"{account_num}에게 {remittance_money}원을 송금받았습니다.")
    @classmethod
    def history(cls):   #내역
        print("-" * 10)
        print("내역")
        print("-" * 10)
        account_num = input("계좌번호: ")
        for account in cls.account_l:
            if account_num == account.account_num:
                for i in account.history_l:
                    print(i)

class UI:
    menu = 0
    @classmethod
    def call(cls):
        print("-" * 30)
        print("1. 계좌생성 | 2. 계좌목록 | 3. 예금 | 4. 출금| 5. 송금| 6. 내역| 7.종료")
        print("-" * 30)
        print("선택> ")
        cls.menu = int(input())
        if cls.menu == 1:
            Account.account_creation()
        elif cls.menu == 2:
            Account.account_list()
        elif cls.menu == 3:
            Account.deposit()
        elif cls.menu == 4:
            Account.whitdraw()
        elif cls.menu == 5:
            Account.remittance()
        elif cls.menu == 6:
            Account.history()
        elif cls.menu == 7:
            print("프로그램이 종료됩니다.")
        else:
            print("1 ~ 7 숫자를 입력하시오")

while True:
    UI.call()
    if UI.menu == 7:
        break
```
