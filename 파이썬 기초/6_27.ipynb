퀵 정렬

```python
def 퀵_정렬(in_l):
    n = len(in_l)
    if n <= 1:
        return in_l
    pivot = in_l[-1]
    g1 = []
    g2 = []
    for i in range(0, n - 1):
        if in_l[i] < pivot:
            g1.append(in_l[i])
        else:
            g2.append(in_l[i])
    return 퀵_정렬(g1) + [pivot] + 퀵_정렬(g2)
```

```python
l = [124, 125, 15, 12, 51, 61, 36, 23, 63, 46, 2, 1643, 165, 6123]
```

```python
end_l = 퀵_정렬(l)
end_l
```

병합 정렬

```python
def 병합_정렬(in_l):
    n = len(in_l)
    if n <= 1:
        return in_l
    mid = n // 2
    g1 = in_l[:mid]
    g2 = in_l[mid:]
    out_l = []
    g1 = 병합_정렬(g1)
    g2 = 병합_정렬(g2)
    while g1 and g2:
        if g1[0] < g2[0]:
            out_l.append(g1.pop(0))
        else:
            out_l.append(g2.pop(0))
    while g1:
        out_l.append(g1.pop(0))
    while g1:
        out_l.append(g2.pop(0))

    return out_l
```

기수 정렬

```python
# 자릿수
# 버킷
# 동작

def 삽입_추출(in_l, d_idx, base = 10):
    buckets = [[] for _ in range(base)]
    for n in in_l:
        d = (n // (base**d_idx)) % base
        buckets[d].append(n)

    out_l = []
    for b in buckets:
        for i in b:
            out_l.append(i)
    return out_l

def 기수_정렬(in_l):
    l = in_l[:]
    max_v = max(l)
    max_d_idx = len(str(max_v))
    for i in range(max_d_idx):
        l = 삽입_추출(l, i, 10)
    return l

l = [124, 125, 15, 12, 51, 61, 36, 23, 63, 46, 2, 1643, 165, 6123]
out_l = 기수_정렬(l)
print(out_l)
```
