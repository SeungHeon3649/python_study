#Q1

```python
x = int(input("값을 입력하시오:"))  # 정수값을 입력 받아서 x에 대입
for i in range(0,x+1,1) :           # 0부터 x + 1 미만까지 1씩증가 반복문
    if i==x:                        # 만약 i == x 이면 (처음 for문 i == 0부터)
        print(i)                    # i 출력
    else:                           # for-else
        print(i,end=",")            # i 출력, 마지막구문 줄바꿈이 아닌 쉼표로 출력
```

#Q2

```python
x = int(input("값을 입력하시오:"))             # 정수값을 입력 받아서 x에 대입
for i in range(1,101,1) :                      # 1부터 100까지 1씩 증가 반복문
    if i<x :                                   # 만약 i < x 이면
        print(i,end=",")                       # i값 출력, end 구문은 ,
    else :                                     # for-else
        print("사용자의 요청에의해 종료")      # "사용자의 요청에 의해 종료" 출력
        break                                  #  반복문 종료
```

#Q3

```python
x = int(input("값을 입력하시오:"))             # 정수값을 입력 받아서 x에 대입
for i in range(1,101) :                        # 1부터 100까지 반복문
    if i%x == 0 :                              # i 나누기 x의 나머지가 0이면
        continue                               # 다음루프로 고
    else :                                     # 아니면
        if i==1:                               # i 가 1이면
            print("1",end="")                  # 1을 출력 end 구문은 공백
        else:                                  # i이 아니면
            print(",",i,sep="",end="")         # ,출력 후 i 출력 간격을 공백 end구문도 공백
```

```python
from random import randint                      # random 패키지에 있는 randint 포함(사용)
x = randint(1,100)                              # randint == 랜덤정수 1에서 100까지 x에 대입

for i in range(100) :                           # 0부터 100까지 반복문
    n = int(input("정답을 입력하시오:"))        # 정답 맞추기 입력 정수형으로, n에 대입
    if n > x :                                  # 만약 n이 x보다 크면
        print("Down")                           # Down 출력
    elif n <x :                                 # 만약 n 보다 x가 크면
        print("Up")                             # UP 출력
    elif n==x :                                 # 만약 n이랑 x랑 같으면
        print(f"{i}회째 정답")                  # i번째 정답 출력
        break                                   # for문 종료
```

#Q6

```python
n = int(input("과목수를 입력하시오:"))    # 정수 입력 후 n에 대입
x=[]                                      # x 빈 리스트 선언
total = 0                                 # total 변수 선언 0 초기화
average = 0                               # average 변수 선언 0 초기화
for i in range(0,n) :                     # 0부터 n까지 반복문
    a=int(input("점수:"))                 # 점수 정수로 입력 a에 대입
    if a < 0:                             # 만약 입력한 정수가 0보다 작으면
        break                             # 반복문 종료
    x.append(a)                           # 아니면 x리스트에 a 추가(뒤에서부터)
    total +=x[-1]                         # 총점합

average = total / len(x)                  # 평균값

print(f"최고 점수 : {max(x)}")            # 최고점수 출력 max함수 이용
print(f"최저 점수 : {min(x)}")            # 최저점수 출력 min함수 이용
print(f"평균 : {average:.2f}")            # 평균 출력
```

```python
n = int(input("행을 입력하시오:"))    # 정수 행 입력 n에 대입
m = int(input("열을 입력하시오:"))    # 정수 열 입력 m에 대입
for u in range(1,n+1) :               # 1부터 n+1까지 반복문
    for o in range(1,m+1) :           # 1부터 m+1까지 반복문
        print(u*o, end=" ")           # u * o 출력 end 구문 스페이스바
    print("\n")                       # 줄바꿈
```
