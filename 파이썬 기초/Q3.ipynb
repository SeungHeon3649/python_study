```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
def 변환함수(data):
  return data.replace('\n','').replace('\t','').replace(',','')
```

```python
url = 'https://finance.naver.com/sise/entryJongmok.naver?&page=1'
html=requests.get(url).text
bs=BeautifulSoup(html,'lxml')
제목=[i.get_text() for i in bs.select('table.type_1 tr')[0].select('th')]
print(제목)
```

```python
url = 'https://finance.naver.com/sise/entryJongmok.naver?&page='
end_data=[제목]
for i in range(1, 21):
    최종정보 = url + str(i)
    html=requests.get(최종정보).text
    bs=BeautifulSoup(html,'lxml')
    bs.get_text()
    for d in bs.select('table.type_1 tr')[1:]:
        if not d.get_text():
            continue
        data=[]
        for i in d.select('td'):
            data.append(변환함수(i.get_text()))
        end_data.append(data)
    time.sleep(1)

df = pd.DataFrame(end_data, columns = 제목)
df.to_csv("data.csv", encoding = 'utf-8')
```
