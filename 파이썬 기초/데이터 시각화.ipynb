```python
x=np.array([1,3,5,7])
plt.bar(x,[1,2,3,4],width=0.5)
plt.bar(x+0.5,[2,3,4,5],width=0.5)
plt.bar(x+1,[3,4,5,6],width=0.5)
plt.xticks(x+0.5,('A','B','C','D'))
plt.show()
```

```python
x=np.array([1,3,5,7])
y=np.array([1,2,3,4])
plt.bar(x,y,width=1,label='data1')
plt.bar(x,y+1,width=1,label='data2',bottom=y)
plt.bar(x,y+2,width=1,label='data3',bottom=y + 1 + y)
plt.legend()
plt.xticks(x,('A','B','C','D'))
plt.show()
```
