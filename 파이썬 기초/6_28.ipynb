```python
import heapq
def 힙_정렬(in_l, 모드 = 1):
    l = in_l[:]
    heapq.heapify(l)
    out_l = []
    for i in range(len(l)):
        if 모드:
            out_l.append(heapq.heappop(l))
        else:
            out_l.insert(0, heapq.heappop(l))
    return out_l
l = [1,7,2,4,5,2,4,7,3,5,8,3,6,4,2,7,6,3,1]
out_l = 힙_정렬(l)
print(out_l)
```

```python
def 삽입정렬(입력데이터_in,모드=0):
  입력데이터 = 입력데이터_in[:]
  정렬결과=[]
  #1. 원소추출
  while 입력데이터:
    추출값 = 입력데이터.pop(0)
    n=len(정렬결과)-1
  #2. 비교
    while True:
      if len(정렬결과)==0 or n==-1:
        비교자리=0
        break
      elif 모드==0 and 정렬결과[n]<추출값:
        비교자리=n+1
        break
      elif 모드==1 and 정렬결과[n]>추출값:
        비교자리=n+1
        break
      else:
        n-=1
  #3. 원소 삽입
    정렬결과.insert(비교자리,추출값)
  return 정렬결과
def 쉘(l,h):
  #분할
  ck_a=[[]for i in range(h)]
  for i in range(len(ck_a)):
    st=i
    add_idx=st
    while add_idx<len(l):
      ck_a[i].append(l[add_idx])
      add_idx+=h
    ck_a[i]=삽입정렬(ck_a[i])
#삽입
  end_l=[None for _ in range(len(l))]
  for i in range(len(ck_a)):
    for j in range(len(ck_a[i])):
      idx=i+(h*j)
      end_l[idx]=ck_a[i][j]
  h//=2
  return  end_l,h

def 쉘_정렬(l,h):
  l,h=쉘(l,h)
  while True:
    if h!=1:
      l,h=쉘(l,h)
    else:
      return 삽입정렬(l)
def 쉘_정렬2(in_l):
  l=in_l[:]
  h=1
  while h<len(l):
    h=3*h+1
  h //=3
  while h>=1:
    for i in range(h,len(l)):
      j=i
      while j>=h and l[j]<l[j-h]:
        l[j],l[j-h]=l[j-h],l[j]
        j-=h
    h//=3
  return l
def 퀵_정렬(in_l):
  n=len(in_l)
  if n<=1:
    return in_l
  pivot=in_l[-1]
  g1=[]
  g2=[]
  for i in range(0,n-1):
    if in_l[i]<pivot:
      g1.append(in_l[i])
    else:
      g2.append(in_l[i])
  return 퀵_정렬(g1) + [pivot] + 퀵_정렬(g2)

def 병합_정렬(in_l):
  n=len(in_l)
  if n<=1:
    return in_l
  mid=n//2
  g1=in_l[:mid]
  g2=in_l[mid:]
  out_l=[]
  g1=병합_정렬(g1)
  g2=병합_정렬(g2)

  while g1 and g2:
    if g1[0] < g2[0]:
      out_l.append(g1.pop(0))
    else:
      out_l.append(g2.pop(0))
  while g1:
    out_l.append(g1.pop(0))
  while g2:
    out_l.append(g2.pop(0))

  return out_l
def 삽입_추출(in_l,d_idx,base=10):
  buckets=[[]for _ in range(base)]
  for n in in_l:
    d=(n // base**d_idx)% base
    buckets[d].append(n)

  out_l=[]
  for b in buckets:
    for i in b:
      out_l.append(i)
  return out_l

def 기수_정렬(in_l):
  l=in_l[:]
  max_v=max(l)
  max_d_idx=len(str(max_v))
  for i in range(max_d_idx):
    l=삽입_추출(l,i,10)
  return l
class 정렬_이진탐색트리:

  class Node:
    def __init__(self,key,L=None,R=None):
      self.key=key
      self.L=L
      self.R=R

  def __init__(self):
    self.root=None
    self.정렬된리스트 = []
  def 삽입(self,key):
    self.root=self.삽입_확인(self.root,key)

  def 삽입_확인(self,n,key):
    if n == None:
      n=self.Node(key)
    elif n.key > key:
      n.L=self.삽입_확인(n.L,key)
    elif n.key < key:
      n.R=self.삽입_확인(n.R,key)
    elif n.key == key:
      pass
      #print(f"{key}값 있다.\n삽입 불가")
    return n

  def 트리_최소값_찾기(self):
    if self.root==None:
      return None
    return self.최소값_노드_찾기(self.root)

  def 최소값_노드_찾기(self,n):
    if n.L==None:
      return n
    return self.최소값_노드_찾기(n.L)

  def 최소값_삭제(self,n):
    if n.L==None:
      return n.R
    else:
      n.L=self.최소값_삭제(n.L)
    return n

  def 삭제(self,key):
    self.root=self.삭제_확인(self.root,key)

  def 삭제_확인(self,n,key):
    if n == None:
      return
    elif n.key == key:
      if n.L==None:
        return n.R
      if n.R==None:
        return n.L
      t=n
      n=self.최소값_노드_찾기(t.R)
      n.R=self.최소값_삭제(t.R)
      n.L=t.L
    elif n.key > key:
      n.L=self.삭제_확인(n.L,key)
    elif n.key < key:
      n.R=self.삭제_확인(n.R,key)
    return n
  def 전위순회(self,n):
    if n!=None:
      #1.출력
      print(n.key,end=" ")
      #2.이동(왼)
      self.전위순회(n.L)
      #3.이동(오)
      self.전위순회(n.R)

  def 중위순회(self,n):
    if n!=None:
      #1.이동(왼)
      self.중위순회(n.L)
      #2.출력
      print(n.key,end=" ")
      #3.이동(오)
      self.중위순회(n.R)

  def 후위순회(self,n):
    if n!=None:
      #1.이동(왼)
      self.후위순회(n.L)
      #2.이동(오)
      self.후위순회(n.R)
      #3.출력
      print(n.key,end=" ")

  def 정렬_중위순회(self,n):
    if n!=None:
      #1.이동(왼)
      self.정렬_중위순회(n.L)
      #2.출력
      #print(n.key,end=" ")
      self.정렬된리스트.append(n.key)
      #print(self.정렬된리스트)
      #3.이동(오)
      self.정렬_중위순회(n.R)
```

```python
class 선형_해시테이블:
    def __init__(self, size):
        self._size = size
        self._data_l = [None for _ in range(size)]
        self._key_l = [None for _ in range(size)]

    def 해시함수(self, key):
        return key % self._size

    def 삽입(self, key, data):
        삽입_주소 = self.해시함수(key)
        죵로확인 = 삽입_주소
        이동수 = 0
        while True:
            if self._key_l[삽입_주소] == None:
                self._key_l[삽입_주소] = key
                self._data_l[삽입_주소] = data
                return
            elif self._key_l[삽입_주소] == key:
                self._data_l[삽입_주소] = data
                return
            이동수 += 1
            삽입_주소 = (삽입_주소 + 이동수) % self._size
            if 종료확인 == 삽입_주소:
                print("가득 참")
                return
    def data_확인(self, key):
        확인_주소 = self.해시함수(key)
        if self._key_l[확인_주소] == None:
            print("삽입된 기록이 없습니다.")
        elif self._key_l[확인_주소] == key:
            return self._data_l[확인_주소]
        else:
            ck_주소 = 확인_주소 + 1
            while ck_주소 != 확인_주소:
                if self._key_l[ck_주소] == key:
                    return self._data_l[확인주소]
                ck_주소 = (ck_주소 + 1) % self._size

    def __str__(self):
        출력문자열 = "\n".join([f"{i:^6}:{self._key_l[i]:^4}:{self._data_l[i]:^10}" for i in range(self._size)])
        return f'{"주소":^4}:{"키":^3}:{"값":^9}\n' + 출력문자열

ht = 선형_해시테이블(10)
ht.삽입(1, "data1")
ht.삽입(2, "data2")
ht.삽입(3, "data3")
ht.삽입(4, "data4")
ht.삽입(5, "data5")
ht.삽입(6, "data6")
ht.삽입(7, "data7")
ht.삽입(8, "data8")
ht.삽입(9, "data9")
ht.삽입(10, "data10")
#print(ht._key_l)
#print(ht._data_l)
print(ht)
```
