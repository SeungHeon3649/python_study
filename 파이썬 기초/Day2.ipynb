```python
from random import randint
cnt = 0
num = []
ques = randint(1, 100)

while True:
  cnt += 1
  ans = int(input(f"{cnt}번째 정답입력 : "))
  if ques == ans:
    break
  elif (ques > ans):
    print("Up")
  else:
    print("Down")
  num.append(ans)

print(f"{cnt}회째 정답!")
print(num)
```

```python
dic = {}
while True:
  Upper = input("입력 : ")
  if not(Upper >= "A" and Upper < "Z"):
    break
  
  if Upper in dic:
    dic[Upper] += 1
  else:
    dic[Upper] = 1

print(dic)
#for key,v in dic.items():
#  print(f"{key} : {v}")
```

```python
dic = {}
Upper_list = list(map(str,input("입력 : ").split()))

for i in range(len(Upper_list)):
  if not(Upper_list[i] >= "A" and Upper_list[i] <= "Z"):
    break
  
  if Upper_list[i] in dic:
    dic[Upper_list[i]] += 1
  else:
    dic[Upper_list[i]] = 1

print(dic)
```

```python
dic = {}
Upper = input()
Upper_list = list(Upper)

for i in range(len(Upper_list)):
  if not(Upper_list[i] >= "A" and Upper_list[i] <= "Z"):
    break
  
  if Upper_list[i] in dic:
    dic[Upper_list[i]] += 1
  else:
    dic[Upper_list[i]] = 1

#print(dic)
for key,v in dic.items():
  print(f"{key} : {v}")
```
