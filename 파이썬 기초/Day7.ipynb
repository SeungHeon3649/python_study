```python
import pickle
import os

class Seat:
    def __init__(self):
        self.name = "  ---  "

    def getname(self):
        return f"""{self.name}"""

    def setname(self, name):
        self.name = name

class Schedule:

    def __init__(self):
        self.s = [Seat() for i in range(8)]

    def getschedule(self, i):
        return self.s[i].getname()

    def setschedule(self, i, name):
        self.s[i].setname(name)

class AirlineBook:

    def __init__(self):
        self.sc = [Schedule() for i in range(3)]

    def getairlinebook(self, i, seat):
        return self.sc[i].getschedule(seat)

    def setairlinebook(self, i, seat, name):
        self.sc[i].setschedule(seat, name)

class UI:
    menu = 0
    schedule = 0
    seat = 0
    name = ""
    s_data1 = ""
    s_data2 = ""
    s_data3 = ""
    air = AirlineBook()

    def __repr__(self):
        for i in range(8):
            self.s_data1 += self.air.getairlinebook(0, i)
            self.s_data2 += self.air.getairlinebook(1, i)
            self.s_data3 += self.air.getairlinebook(2, i)
        return f"""{self.s_data1}
{self.s_data2}
{self.s_data3}
"""

    def reservation(self): #예약
        for i in range(8):
            print(self.air.getairlinebook(self.schedule - 1, i), end = "")
        print("\n좌석 번호 >> ", end = "")
        self.seat = int(input())
        print("이름 입력 >> ", end = "")
        self.name = input()
        self.air.setairlinebook(self.schedule - 1, self.seat- 1, self.name)

    def cancellation(self): #취소
        for i in range(8):
            print(self.air.getairlinebook(self.schedule - 1, i), end = "")
        print("\n좌석 번호 >> ", end = "")
        self.seat = int(input())
        print("이름 입력 >> ", end = "")
        self.name = input()
        self.air.setairlinebook(self.schedule - 1, self.seat - 1, "  ---  ")

    def print_airlinebook(self, num):
        for i in range(8):
            print(self.air.getairlinebook(num, i), end = "")
        print("")

    def call(self):
        print("-" * 30)
        print("1. 예약 | 2. 취소 | 3. 보기 | 4. 끝내기 >> ", end = "")
        self.menu = int(input())
        if self.menu == 1:
            print("1. 07시 | 2. 12시 | 3. 17시 >> ", end = "")
            self.schedule = int(input())
            if self.schedule == 1:
                print("07시 : ", end = "")
                self.reservation()
            elif self.schedule == 2:
                print("12시 : ", end = "")
                self.reservation()
            elif self.schedule == 3:
                print("17시 : ", end = "")
                self.reservation()
        elif self.menu == 2:
            print("1. 07시 | 2. 12시 | 3. 17시 >> ", end = "")
            self.schedule = int(input())
            if self.schedule == 1:
                print("07시 : ", end = "")
                self.cancellation()
            elif self.schedule == 2:
                print("12시 : ", end = "")
                self.cancellation()
            elif self.schedule == 3:
                print("17시 : ", end = "")
                self.cancellation()
        elif self.menu == 3:
            print("07시 : ", end = "")
            self.print_airlinebook(0)
            print("12시 : ", end = "")
            self.print_airlinebook(1)
            print("17시 : ", end = "")
            self.print_airlinebook(2)
        elif self.menu == 4:
            print("프로그램이 종료됩니다.")
        else:
            print("1 ~ 4 숫자를 입력하시오.")

def main():
    ui = UI()
    while True:
        ui.call()
        if ui.menu == 4:
            with open("management_data.p",'wb') as f:
                pickle.dump(ui, f)
            print(pickle.load(open("management_data.p",'rb')))
            break

if __name__ == "__main__":
    try:
        pickle.load(open("management_data.p",'rb'))
    except:
        pass
    main()
```

```python
class 계좌관리:
  총계좌정보={}
  def __init__(self,고객명,계좌번호,금액):
    self.고객명=고객명
    self.계좌번호=계좌번호
    self.금액=금액
  def __str__(self):
    return f"|{self.계좌번호:^11}|{self.고객명:^5}|{self.금액:>10}|"
  @classmethod
  def 생성(cls):
    print(f"""--------------
계좌생성
--------------""")
    계좌번호=input("계좌번호: ")
    고객명=input("계좌주: ")
    금액=int(input("초기입금액: "))
    n_통장=cls(고객명,계좌번호,금액)
    if 계좌번호 in cls.총계좌정보:
      print("중복된 계좌 번호 입니다.")
      return
    cls.총계좌정보[계좌번호]=n_통장
    print("결과: 계좌가 생성되었습니다.")

  @classmethod
  def 출력(cls):
    print(f"""--------------
계좌목록
--------------""")
    for 통장 in cls.총계좌정보.values():
      print(통장)
  @classmethod
  def 입금(cls):
    print(f"""--------------
예금
--------------""")
    입금계좌=input("계좌번호: ")
    입금액=int(input("예금액: "))
    if 입금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    호출통장=cls.총계좌정보[입금계좌]
    호출통장.금액+=입금액
    print('결과: 예금이 성공되었습니다.')

  @classmethod
  def 출금(cls):
    print(f"""--------------
출금
--------------""")
    출금계좌=input("계좌번호: ")
    출금액=int(input("금액: "))
    if 출금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    호출통장=cls.총계좌정보[출금계좌]
    if 호출통장.금액<출금액:
      print("잔액 부족")
      return
    호출통장.금액-=출금액
    print('결과: 출금이 성공되었습니다.')

  @classmethod
  def 종료(cls):
     print("프로그램 종료")
     return True
```

```python
class 계좌관리:
  총계좌정보={}
  def __init__(self,고객명,계좌번호,금액):
    self.고객명=고객명
    self.계좌번호=계좌번호
    self.금액=금액
    self.통장내역=f'{"초기금액:"+str(금액):10s}|{"합계금액:"+str(금액):10s}\n'

  def __str__(self):
    return f"|{self.계좌번호:^11}|{self.고객명:^5}|{self.금액:>10}|"
  def 내역기록(self,금액):
    if 금액>=0:
      self.통장내역 += f'{"입금  액:"+str(금액):10s}|{"합계금액:"+str(self.금액):10s}\n'
    else:
      self.통장내역 += f'{"출금  액:"+str(금액* -1):10s}|{"합계금액:"+str(self.금액):10s}\n'

  @classmethod
  def 생성(cls):
    print(f"""--------------
계좌생성
--------------""")
    계좌번호=input("계좌번호: ")
    고객명=input("계좌주: ")
    금액=int(input("초기입금액: "))
    n_통장=cls(고객명,계좌번호,금액)
    if 계좌번호 in cls.총계좌정보:
      print("중복된 계좌 번호 입니다.")
      return
    cls.총계좌정보[계좌번호]=n_통장
    print("결과: 계좌가 생성되었습니다.")
  @classmethod
  def 출력(cls):
    print(f"""--------------
계좌목록
--------------""")
    for 통장 in cls.총계좌정보.values():
      print(통장)
  @classmethod
  def 입금(cls):
    print(f"""--------------
예금
--------------""")
    입금계좌=input("계좌번호: ")
    입금액=int(input("예금액: "))
    if 입금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    호출통장=cls.총계좌정보[입금계좌]
    호출통장.금액+=입금액
    호출통장.내역기록(입금액)
    print('결과: 예금이 성공되었습니다.')
  @classmethod
  def 출금(cls):
    print(f"""--------------
출금
--------------""")
    출금계좌=input("계좌번호: ")
    출금액=int(input("금액: "))
    if 출금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    호출통장=cls.총계좌정보[출금계좌]
    if 호출통장.금액<출금액:
      print("잔액 부족")
      return
    호출통장.금액-=출금액
    출금액*=-1
    호출통장.내역기록(출금액)
    print('결과: 출금이 성공되었습니다.')
  @classmethod
  def 종료(cls):
     print("프로그램 종료")
     return True
  @classmethod
  def 내역_조회(cls):
    print(f"""--------------
내역조회
--------------""")
    확인계좌=input("계좌번호: ")
    if 확인계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    print(f"{확인계좌}의 내역")
    print(cls.총계좌정보[확인계좌].통장내역)
    print("내역 출력이 완료되었습니다.")
  @classmethod
  def 송금(cls):
    print(f"""--------------
송금
--------------""")
    출금계좌=input("출금 계좌번호: ")
    if 출금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    입금계좌=input("입금 계좌번호: ")
    if 입금계좌 not in cls.총계좌정보:
      print("없는 계좌 번호 입니다.")
      return
    출금액=int(input("송금액: "))
    출금통장=cls.총계좌정보[출금계좌]
    입금통장=cls.총계좌정보[입금계좌]
    if 출금통장.금액<출금액:
      print("잔액 부족")
      return
    출금통장.금액-=출금액
    입금액=출금액
    출금액*=-1
    출금통장.내역기록(출금액)
    입금통장.금액+=입금액
    입금통장.내역기록(입금액)
    print("송금 완료")
```
