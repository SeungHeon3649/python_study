연결리스트 구현

```python
class Node:#클래스 생성 클래스 이름 Node
  def __init__(self,data,link):#클래스 생성자(반드시 입력 2개 받는) 생성
    self.data=data#인스턴스 필드 data 매개변수 data 받으면서 생성
    self.next=link#인스턴스 필드 next 매개변수 link 받으면서 생성

class 연결리스트:#클래스 생성 클래스 이름 연결리스트
  def __init__(self):#클래스 생성자(입력없는) 생성
    self.__head=None#인스턴스 필드 __head 생성 및 None로 초기화
    self.__size=0#인스턴스 필드 __size 생성 및 0로 초기화

  def l_size(self):#인스턴스 메소드 l_size(입력 없는)생성
    return self.__size==0#s인스턴스 필드 __size ==0 결과 리턴
  def get_size(self):#인스턴스 메소드 get_size(입력 없는)생성
    return self.__size#s인스턴스 필드 __size 리턴
  def __str__(self):#s인스턴스 메소드 __str__ 오버라이딩
    출력문자열=''#지역 변수 출력문자열 ''으로 초기화
    p=self.__head#지역변수 p를 이용한 인스턴스 필드 __head 접근 경로생성
    while p:#지역변수 p를 이용한 반복문 설정
      출력문자열+=f'{p.data} ->'#출력문자열에 리스트 data 문자열 추가 연산
      p=p.next#지역변수 p를 이용한 노드 이동
    return 출력문자열[:-2]#끝 두글자를 생략한 출력문자열 리턴

  #끝-중간 삽입(기본동작은 끝에서 삽입)
  def in_sert(self,x,n=-1):#n=인덱스 #인스턴스 메소드 in_sert(1~2개입력 1개 기본값)생성
    if self.__size < n:#self.__size 값이 입력 n보다 작으면 동작 조건문 생성
      print("범위 초과로 삽입 불가")#범위 초과로 삽입 불가 출력
      return #삽입 동작 종료

    if self.l_size():#리스트에 데이터 없을때 삽입
      self.__head=Node(x,self.__head)#머리에서 삽입
    elif n==-1:#끝에서 삽입
      p=self.__head#지역변수 p를 이용한 인스턴스 필드 __head 접근 경로생성
      while p.next:#p를 이용한 반복문 생성
        p=p.next#다음 노드 이동
      else:#최종노드를 찾을때 동작
        p.next=Node(x,None)#최종 노드 뒤 생성 삽입
    else:#중간 삽입
      p=self.__head#지역변수 p를 이용한 인스턴스 필드 __head 접근 경로생성
      ck=0#동작 확인 목적 변수 생성
      while n:#입력된 숫자의 값이 n==0일때 정지하는 반북문 생성
        if ck==n-1:#정지 목적 조건 생성
          break#반복 종료
        p=p.next#다음노드 이동
        ck+=1#이동횟수기록
      else:#n==0일때만 동작하는 구문 설계
        self.__head=Node(x,self.__head)#머리에서 삽입
        self.__size+=1#원소갯수 1개 증가 기록
        return#삽입 함수 종료
      p.next=Node(x,p.next)#break 로 탈출시 동작하는 위치에 원소 삽입
    self.__size+=1#원소 갯수 1개 증가 기록
  #끝-중간 삭제(기본동작은 끝에서 삽입)
  def del_l(self,n=-1):#인스턴스 메소드 del_l(1개입력 1개 기본값)생성
    if self.l_size() or n>self.__size:#리스트에 원소가 없거나 가지고 있는 원소보다 큰 인덱스를 통해 접근 시 조건 설계
      print("삭제 연산은 원소가 있을때만 동작 해야 한다.")#출력
      return#삭제 함수 종료
    #1.머리에서 삭제
    if n==0:#머리에서 삭제
      self.__head=self.__head.next#head 필드값을 head.next노드의 주소값으로 교채
      print("머리에서 삭제")#출력
    #2.끝에서 삭제
    elif n==-1:#끝에서 삭제
      e=self.__head#지역변수 e를 이용한 인스턴스 필드 __head 접근 경로생성
      while e.next:#최종 노드 발견시 종료
        t=e#최종 노드 전 노드 기록
        e=e.next#최종 노드 찾기 이동
      t.next=None#최종 노드 전 노드를 통한 최종노드 제거
      print("끝에서 삭제")#출력
    #3. 중간에서 삭제 n!=0,n!=-1,1<=n<=size
    else:#중간에서 삭제
      p=self.__head#지역변수 p를 이용한 인스턴스 필드 __head 접근 경로생성
      for i in range(n):#n값을 이용한 지정 위치 노드 접근
        t=p#이동 전 노드 기록
        p=p.next#이동
      t.next=p.next#찾은 노드 를 이용한 노드 삭제
      print("중간에서 삭제")#출력
```

```python
class 연결리스트:
  class Node:
    def __init__(self,data,link):
      self.data=data
      self.next=link
    def __repr__(self):
      return f'data:[{self.data}]\nlink:[{self.next}]'

  def __init__(self,*x):
    self.__head=None
    self.__size=0
    if x:
      for i in x:
        self.삽입연산(i)
      else:
        print("입력값 있는 생성")
    else:
      print("빈 리스트 생성")
    self.p=self.__head

  #개별리턴
  def __iter__(self):
    return self
  def __next__(self):
    if self.p==None:
      self.p=self.__head
      raise  StopIteration
    out_data=self.p.data
    self.p=self.p.next
    return out_data

  #리스트 출력
  def __repr__(self):
    출력문자열=' '
    p=self.__head
    while p:
      출력문자열+=f'{p.data} -> '
      p=p.next
    return '['+출력문자열[:-3]+']'

  def idx_찾기(self,idx):
    if idx>self.__size:
      print("초과 호출")
      return None
    #머리 찾기
    elif idx==0:
      return self.__head
    elif idx==self.__size:
      print("노드가 없습니다.")
      return None
    p=self.__head
    for i in range(idx):
      p=p.next
    else:
      return p

  def 삽입연산(self,x,n=-1):
    #0. 범위 확인
    if self.__size <n:
      print("범위 초과 삽입 불가")
      return
    #1. 빈리스트 삽입 - 머리에서 삽입
    elif self.__size==0 or n==0:
      self.__head=self.Node(x,self.__head)
      print("머리에서 삽입")
    #2. 끝에서 삽입
    elif n==-1:
      p=self.idx_찾기(self.__size-1)
      p.next=self.Node(x,None)
      print("꼬리에서 삽입")
    #3. 중간에서 삽입
    else:
      p=self.idx_찾기(n-1)
      p.next=self.Node(x,p.next)
    self.__size+=1
    self.p=self.__head

  def 삭제연산(self,n=-1):
    if self.__size==0 or n>=self.__size:
      print("삭제연산은 원소가 있을때만 동작 해야 한다.")
      return
    #1.머리
    if n==0:
      self.__head=self.__head.next
      print("머리에서 삭제")
    #2.끝
    elif n==-1:
      t=self.idx_찾기(self.__size-2)
      t.next=None
      print("끝에서 삭제")
    #3.중간
    else:
      t=self.idx_찾기(n-1)
      p=t.next
      t.next=p.next
      print("중간에서 삭제")
    self.__size-=1
    self.p=self.__head
```

```python
class 원형연결리스트:
  class Node:
    def __init__(self,data,link):
      self.data=data
      self.next=link
  def __init__(self,*x):
    self.__head=None
    self.__size=0
    if x:
      for i in x:
        self.삽입연산(i)
      print("데이터를 입력하면서 생성")
    else:
      print("빈 리스트 생성")
  #리스트 출력
  def __repr__(self):
    if self.__size==0:
      return "[]"
    출력문자열=' -> '
    p=self.__head
    for i in range(self.__size):
      출력문자열+=f'{p.data} -> '
      p=p.next
    return '['+출력문자열+']'

  def idx_찾기(self,idx):
    if idx>=self.__size:
      print("초과 호출")
      return None
    #머리 찾기
    elif idx==0:
      return self.__head
    elif idx==-2:
      l=self.__head
      while l.next!=self.__head:
        l=l.next
      return l
    p=self.__head
    for i in range(idx):
      p=p.next
    else:
      return p

  def 삽입연산(self,x,n=-1):
    #0. 범위 확인
    if self.__size <n:
      print("범위 초과 삽입 불가")
      return
    #1. 빈리스트 삽입
    elif self.__size==0:
      self.__head=self.Node(x,None)
      self.__head.next=self.__head
      print("빈 리스트에서 삽입")
    #2. 중간에서 삽입
    else:
      p=self.idx_찾기(n-1)
      p.next=self.Node(x,p.next)
    self.__size+=1

  def 삭제연산(self,n=-1):
    if self.__size==0 or n>=self.__size:
      print("삭제연산은 원소가 있을때만 동작 해야 한다.")
      return
    #1.머리
    if n==0:
      del_n=self.__head
      self.__head=self.__head.next
      l=self.__head
      while l.next!=del_n:
        l=l.next
      l.next=self.__head
      print("머리에서 삭제")
    #2.끝
    elif n==-1:
      t=self.idx_찾기(self.__size-2)
      t.next=self.__head
      print("끝에서 삭제")
    #3.중간
    else:
      t=self.idx_찾기(n-1)
      p=t.next
      t.next=p.next
      print("중간에서 삭제")
    self.__size-=1
    self.p=self.__head
```

```python
class 원형연결리스트_L:
  class Node:
    def __init__(self,data,link):
      self.data=data
      self.next=link
  def __init__(self,*x):
    self.__last=None
    self.__size=0
    if x:
      for i in x:
        self.삽입연산(i)
      print("데이터를 입력하면서 생성")
    else:
      print("빈 리스트 생성")
  #리스트 출력
  def __repr__(self):
    if self.__size==0:
      return "[]"
    출력문자열=' -> '
    p=self.__last.next
    for i in range(self.__size):
      출력문자열+=f'{p.data} -> '
      p=p.next
    return '['+출력문자열+']'

  def idx_찾기(self,idx):
    #머리 찾기
    if idx==0:
      return self.__last.next
    elif idx==-1:
      return self.__last
    p=self.__last#self.__head
    for i in range(idx):
      p=p.next
    else:
      return p

  def 삽입연산(self,x,n=-1):
    #0. 범위 확인
    if self.__size <=n:
      print("범위 초과 삽입 불가")
      return
    #1. 빈리스트 삽입
    elif self.__size==0:
      self.__last=self.Node(x,None)
      self.__last.next=self.__last
      print("빈 리스트에서 삽입")
    #2. 중간에서 삽입
    else:
      p=self.idx_찾기(n)
      p.next=self.Node(x,p.next)
      if n==-1:
        self.__last=p.next
    self.__size+=1

  def 삭제연산(self,n=-1):
    if self.__size==0 or n>=self.__size:
      print("삭제연산은 원소가 있을때만 동작 해야 한다.")
      return
    #1.머리
    if n==0:
      del_n=self.__last.next
      self.__last.next=del_n.next
      print("머리에서 삭제")
    #2.끝
    elif n==-1:
      t=self.idx_찾기(self.__size-1)
      t.next=self.__last.next
      print("끝에서 삭제")
    #3.중간
    else:
      t=self.idx_찾기(n)
      p=t.next
      t.next=p.next
      print("중간에서 삭제")
    self.__size-=1
```

이중연결 리스트

```python
class 이중_연결리스트:
  class Node:
    def __init__(self,data,rlink,llink):
      self.data=data
      self.정방향=rlink
      self.역방향=llink
    def __repr__(self):
      return f'역방향:[{self.역방향}]|data:[{self.data}]|정방향:[{self.정방향}]'

  def __init__(self):
    self.__head=self.Node(None,None,None)
    self.__last=self.Node(None,None,self.__head)
    self.__head.정방향=self.__last
    self.__size=0

  #리스트 출력
  def __repr__(self):
    출력문자열=' '
    p=self.__head
    while p:
      출력문자열+=f'{p.data} <-> '
      p=p.정방향
    return '['+출력문자열[:-4]+']'
  #머리
  def 정방향_삽입연산(self,입력데이터):
    머리=self.__head
    nn=self.Node(입력데이터,머리.정방향,머리)
    머리.정방향.역방향=nn
    머리.정방향=nn
    self.__size+=1
    print("정방향삽입(머리삽입)")
  def 정방향_삭제연산(self):
    if self.__size==0:
      print("원소가 없습니다.")
      return
    del_n=self.__head.정방향
    self.__head.정방향=del_n.정방향
    del_n.정방향.역방향=self.__head
    self.__size-=1
    print("정방향삭제(머리삭제)")
    return del_n.data
  #꼬리
  def 역방향_삽입연산(self,입력데이터):
    꼬리=self.__last
    nn=self.Node(입력데이터,꼬리,꼬리.역방향)
    꼬리.역방향.정방향=nn
    꼬리.역방향=nn
    self.__size+=1
    print("역방향삽입(꼬리삽입)")
  def 역방향_삭제연산(self):
    if self.__size==0:
      print("원소가 없습니다.")
      return
    del_n=self.__last.역방향
    self.__last.역방향=del_n.역방향
    del_n.역방향.정방향=self.__last
    self.__size-=1
    print("역방향삭제(꼬리삭제)")
    return del_n.data
```

```python
a=이중_연결리스트()
a.정방향_삽입연산(10)
a.정방향_삽입연산(20)
a.정방향_삽입연산(30)
a.정방향_삽입연산(40)
a.역방향_삽입연산(10)
a.역방향_삽입연산(20)
a.역방향_삽입연산(30)
a.역방향_삽입연산(40)
print(a)
```
