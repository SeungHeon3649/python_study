```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
def 변환함수(data):
  return data.replace('\n','').replace('\t','').replace(',','')
```

```python
url = "https://finance.naver.com/marketindex/worldDailyQuote.naver?fdtc=4&marketindexCd=FX_GBPUSD&page=1"

# HTTP GET 요청
response = requests.get(url)

# 요청이 성공했을 경우
if response.status_code == 200:
    # HTML 파싱
    soup = BeautifulSoup(response.content, "html.parser")

    # 환율 데이터 추출
    table = soup.find("table", class_="tbl_exchange")
    rows = table.find_all("tr")

    for row in rows:
        columns = row.find_all("td")
        if len(columns) == 4:
            date = columns[0].text.strip()
            exchange_rate = columns[1].text.strip()
            compared_day = columns[2].text.strip()
            fluctuation_rate = columns[3].text.strip()
            print(f"날짜: {date}, 환율: {exchange_rate}, 전일대비: {compared_day}, 등락율: {fluctuation_rate}")
else:
    print("요청에 실패했습니다.")
```

```python
#제목 = ['날짜', '종가', '전일대비', '등락율']
url = 'https://finance.naver.com/marketindex/worldDailyQuote.naver?fdtc=4&marketindexCd=FX_GBPUSD&page='
end_data=[]
for i in range(1, 21):
    최종정보 = url + str(i)
    html=requests.get(최종정보).text
    bs=BeautifulSoup(html,'lxml') # text파일을 변환
    bs.get_text()
    for d in bs.select('table.tbl_exchange tr')[0:]:
        if not d.get_text():
            continue
        data=[]
        for i in d.select('td'):
        end_data.append(data)
    time.sleep(1)

df = pd.DataFrame(end_data, columns = 제목)
df.to_csv("data.csv", encoding = 'utf-8')
```
