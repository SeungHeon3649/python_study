```python
student = np.random.randint(0,101,100)
student_mean = np.mean(student)
if student_mean >= 50:
    print("전원 통과")
if len(student[student >= 50]) >= 50:
    print("학년1등반")
if len(student[student == 100]) >= 1:
    print("전교1등반")
```

```python
student = np.random.randint(0,101,100)
student_mean = np.mean(student)
if student_mean >= 50:
    print("전원 통과")
if np.all(student >= 50):
    print("학년1등반")
if np.any(student == 100):
    print("전교1등반")
```
