끝에서 삽입

```python
def 리스트_출력(st):
  print('[',end=" ")
  while st:
    if not st.next:
      print(st.data,end=" ")
    else:
      print(st.data,end=" -> ")
    st=st.next
  else:
    print("]")
    print("리스트의 출력이 종료 되었습니다.")
def 끝에서_삽입(st,x):
  #빈 리스트 확인
  if not st:
    st=Node(x,None)
  else:
    nn=st
    while nn.next:
      nn=nn.next
    else:
      nn.next=Node(x,None)
  return st
def 중간_삽입(st,n,x):
  #비어있는 삽입
  if not st:
    st=Node(x,None)
    return st
  #머리에서 삽입
  if n==1:
    st=Node(x,st)
    return st
  #n번째 노드 위치에서 삽입(n-1의 위치 노드)
  p=st#1번째 노드
  이동수=0
  #삽입전 노드위치 찾기
  while True:
    #이동 종료
    if 이동수==n-2:
      break
    p=p.next
    이동수+=1
  #1.데이터를 가지고있는 노드 생성
  #next_n=Node(x)
  #2.전 노드의 link연결
  #next_n.next=p.next
  #3.생성노드와 전노드 연결
  #p.next=next_n
  p.next=Node(x,p.next)
  return st
```

삭제
  - 빈곳 - 절대 불가
  - 끝 - 끝노드 제거
  - 머리(중간) - 삭제 노드 뒷노드 앞단 연결

```python
def 끝_삭제(st):
  if st==None:
    print("빈 리스트는 삭제 불가")
    return
  t_p=st
  while t_p.next.next:
    t_p=t_p.next
  t_p.next=None
  return st
def 중간_삭제(st,n):
  if st==None:
    print("빈 리스트는 삭제 불가")
    return
  if n==0:
    st=st.next
    return st
  else:
    t_p=st#1번째 노드
    if n==1:
      st.next=st.next.next
    else:
      for i in range(n-1):#2
        t_p=t_p.next
      t_p.next=t_p.next.next
  return st
```

```python
st=None
리스트_출력(st)
st=끝에서_삽입(st,10)
리스트_출력(st)
st=끝에서_삽입(st,20)
리스트_출력(st)
st=끝에서_삽입(st,30)
리스트_출력(st)
```

연결 리스트 구현

```python
class Node:
  def __init__(self,data,link):
    self.data=data
    self.next=link

class 연결리스트:
  def __init__(self):
    self.__head=None
    self.__size=0

  def l_size(self):
    return self.__size==0
  def get_size(self):
    return self.__size
  def __str__(self):
    출력문자열=''
    p=self.__head
    while p:
      출력문자열+=f'{p.data} ->'
      p=p.next
    return 출력문자열[:-2]

  #끝-중간 삽입(기본동작은 끝에서 삽입)
  def in_sert(self,x,n=-1):#n=인덱스
    if self.__size < n:
      print("범위 초과로 삽입 불가")
      return
    if self.l_size():#없을때 삽입
      self.__head=Node(x,self.__head)
    elif n==-1:#끝에서 삽입
      p=self.__head
      while p.next:
        p=p.next#이동
      else:
        p.next=Node(x,None)
    else:
      p=self.__head
      ck=0
      while n:
        if ck==n-1:
          break
        p=p.next
        ck+=1
      else:
        self.__head=Node(x,self.__head)
        self.__size+=1
        return
      p.next=Node(x,p.next)
    self.__size+=1
  #끝-중간 삭제(기본동작은 끝에서 삽입)
  def del_l(self,n=-1):
    if self.l_size() or n>self.__size:
      print("삭제 연산은 원소가 있을때만 동작 해야 한다.")
      return
    #1.머리에서 삭제
    if n==0:
      self.__head=self.__head.next
      print("머리에서 삭제")
    #2.끝에서 삭제
    elif n==-1:
      e=self.__head
      while e.next:
        t=e
        e=e.next
      t.next=None
      print("끝에서 삭제")
    #3. 중간에서 삭제 n!=0,n!=-1,1<=n<=size
    else:
      p=self.__head
      for i in range(n):
        t=p
        p=p.next
      t.next=p.next
      print("중간에서 삭제")
```

```python
l_l=연결리스트()
print(l_l.get_size())
l_l.in_sert(10)
l_l.in_sert(20)
l_l.in_sert(30)
l_l.in_sert(100,0)
l_l.in_sert(100,4)
l_l.in_sert(50,3)
print(l_l)
l_l.get_size()
print(l_l)
```
