```python
def 삽입정렬(입력데이터,모드=0):
  정렬결과=[]
  #1. 원소추출
  while 입력데이터:
    추출값 = 입력데이터.pop(0)
    n=len(정렬결과)-1
    print(n)
  #2. 비교
    while True:
      if len(정렬결과)==0 or n==-1:
        비교자리=0
        break
      elif 모드==0 and 정렬결과[n]<추출값:
        비교자리=n+1
        break
      elif 모드==1 and 정렬결과[n]>추출값:
        비교자리=n+1
        break
      else:
        n-=1
  #3. 원소 삽입
    정렬결과.insert(비교자리,추출값)
  return 정렬결과
삽입정렬([8,3,5,1,2,0,4,23])
```

```python
def 쉘(l,h):
  #분할
  ck_a=[[]for i in range(h)]
  for i in range(len(ck_a)):
    st=i
    add_idx=st
    while add_idx<len(l):
      ck_a[i].append(l[add_idx])
      add_idx+=h
    ck_a[i]=삽입정렬(ck_a[i])
#삽입
  end_l=[None for _ in range(len(l))]
  for i in range(len(ck_a)):
    for j in range(len(ck_a[i])):
      idx=i+(h*j)
      end_l[idx]=ck_a[i][j]
  h//=2
  return  end_l,h

def 쉘_정렬(l,h):
  l,h=쉘(l,h)
  while True:
    if h!=1:
      l,h=쉘(l,h)
    else:
      return 삽입정렬(l)
l=[1,45,2,5,2,3,45,1]
h=4
쉘_정렬(l,h)
```

```python
from random import randint,sample
import time
import matplotlib.pyplot as plt
```

```python
def 쉘_정렬(in_l):
  l=in_l[:]
  h=1
  while h<len(l):
    h=3*h+1
  h //=3
  while h>=1:
    plt.bar(range(len(l)),l,width=1)
    plt.show()
    for i in range(h,len(l)):
      j=i
      while j>=h and l[j]<l[j-h]:
        l[j],l[j-h]=l[j-h],l[j]
        j-=h
    h//=3
  return l

l=[i for i in range(10,-1,-1)]
쉘_정렬(l)
```

```python
l=[1,3,4,2,5,2,65,23,4,76,23,4,2]
def 퀵_정렬(in_l):
  n=len(in_l)
  if n<=1:
    return in_l
  pivot=in_l[-1]
  g1=[]
  g2=[]
  for i in range(0,n-1):
    if in_l[i]<pivot:
      g1.append(in_l[i])
    else:
      g2.append(in_l[i])
  return 퀵_정렬(g1) + [pivot] + 퀵_정렬(g2)

퀵_정렬(l)
```

```python
l=[1,3,4,2,5,2,65,23,4,76,23,4,2]
def 병합_정렬(in_l):
  n=len(in_l)
  if n<=1:
    return in_l
  mid=n//2
  g1=in_l[:mid]
  g2=in_l[mid:]
  out_l=[]
  g1=병합_정렬(g1)
  g2=병합_정렬(g2)

  while g1 and g2:
    if g1[0] < g2[0]:
      out_l.append(g1.pop(0))
    else:
      out_l.append(g2.pop(0))
  while g1:
    out_l.append(g1.pop(0))
  while g2:
    out_l.append(g2.pop(0))

  return out_l
병합_정렬(l)
```

```python
def 삽입_추출(in_l,d_idx,base=10):
  buckets=[[]for _ in range(base)]
  for n in in_l:
    d=(n // base**d_idx)% base
    buckets[d].append(n)

  out_l=[]
  for b in buckets:
    for i in b:
      out_l.append(i)
  return out_l

def 기수_정렬(in_l):
  l=in_l[:]
  max_v=max(l)
  max_d_idx=len(str(max_v))
  for i in range(max_d_idx):
    l=삽입_추출(l,i,10)
  return l

l=[1,5,23,5,2,6,23,34,6,2,3]
기수_정렬(l)

#자리수
#버킷
#동작
```
