```python
x = np.array([1,3,5,7])
plt.bar(x, [1,2,3,4], width = 1, label = 'data1')
plt.bar(x, [2,3,4,5], width = 1, label = 'data2', bottom = [1,2,3,4])
plt.bar(x,[3,4,5,6], width = 1, label = 'data3', bottom = [3,5,7,9])
plt.legend()
plt.xticks(x, ('A', 'B', 'C', 'D'))
plt.show()
```

```python
#선형회귀 추세선 데이터 기반으로 데이터의 x값 대비 y값의 변화를 예측하는 직선
# 점 = 산점도, 직선 = 회귀 직선, 색칠 = 신뢰구간
sns.regplot(df1, x = 'total_bill', y = 'tip')
```

```python
# 산점도
sns.scatterplot(df1, x = 'total_bill', y = 'tip', hue = 'sex')
```

```python
# 막대 그래프
# y값이 연속형데이터일 경우 해당값들의 평균 값 나타냄
# 검은선 = 신뢰구간
sns.barplot(df1, x = 'sex', y = 'tip', hue = 'smoker')
```

밑부터는 범주형 데이터에 유용하게 쓸 수 있음

```python
# 바이올린 플롯
# x 축에는 범주형 데이터, y 축에는 연속형 데이터
sns.violinplot(df1, x = 'sex', y = 'tip', hue = 'smoker')
```

```python
# 바이올린과 형태가 같지만 산점도를 이용하여 데이터 분포를 나타냄
sns.swarmplot(df1, x = 'sex', y = 'tip', hue = 'smoker')
```
