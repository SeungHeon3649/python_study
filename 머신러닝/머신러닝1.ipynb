1. 데이터 수집

```python
입력=[[270,'울퉁불퉁'],[250,'울퉁불퉁'],[220,'매끈'],[240,'매끈']]
결과=[["오렌지"],['오렌지'],['사과'],['사과']]
```

2. 데이터 분석 및 준비

```python
#울퉁붕퉁=>0/매끈=1
#오렌지=0,사과=1
입력=[[270,0],[250,0],[220,1],[240,1]]
결과=[[0],[0],[1],[1]]
```

3. 학습

```python
m = DecisionTreeClassifier()
m.fit(입력,결과)
```

```python
m.score(입력,결과)
```

```python
m.predict([[260,0]])
```
